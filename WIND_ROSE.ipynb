```python
import numpy as np
import matplotlib.pyplot as plt
import xarray as xr
from windrose import WindroseAxes

# Load wind speed and wind direction data
data = xr.open_dataset('Nkawkaw_wind_speed_wind_direction.nc')
wind_speed = data['WS10M'].values.flatten()
wind_direction = data['WD10M'].values.flatten()

# Calculate percentage values for wind directions
total_samples = len(wind_direction)
direction_counts, _ = np.histogram(wind_direction, bins=np.arange(0, 361, 45))
direction_percentages = direction_counts / total_samples * 100

# Create a WindroseAxes object
ax = WindroseAxes.from_ax()

# Set the face color for the WindroseAxes object
ax.set_facecolor('lavender')

# Plot the wind rose with customized appearance
ax.bar(wind_direction, wind_speed, edgecolor='white', nsector=16, cmap=plt.cm.plasma)

# Annotate percentages of wind directions
angles = np.linspace(0, 2 * np.pi, 16, endpoint=False)
for angle, percentage in zip(angles, direction_percentages):
    ax.text(angle, 70, f'{percentage:.1f}%', fontsize='x-large', ha='right', va='top')

# Set legend and adjust its height
legend = ax.set_legend(title='Wind Speed (m/s)', fontsize='large')
legend.set_bbox_to_anchor((1.15, 0.7))

# Customize plot appearance
plt.title('Wind Rose Plot for Nkawkaw', fontsize='x-large', fontweight='bold', color='blue')
ax.set_xlabel('Wind Direction (°)', fontsize='large', color='blue')
ax.set_ylabel('Frequency', fontsize='large', labelpad=50, color='blue')
ax.tick_params(labelsize='large', color='blue')

# Add grid lines
ax.grid(True, linestyle='--', linewidth=1.5, alpha=0.5, color='blue')

# Show the plot
plt.show()
```
